# Import Libraries

```python
import pandas as pd
import numpy as np
import os
```

# Global Variable

```python
df = None
```

# Load Excel file

```python
def load_file():
    global df
    file_path = input("Enter Excel file path: ")
    
    try:
        df = pd.read_excel(file_path)
        print("File loaded successfully!")
        print("Shape:", df.shape)
    except FileNotFoundError:
        print("Error: File not found.")
    except Exception as e:
        print("Error:", e)
```

# Standardize Coloumn Names

```python
def standardize_columns():
    global df
    if df is not None:
        df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
        print("Column names standardized.")
    else:
        print("Load data first.")
```

# Remove Duplicates

```python
def remove_duplicates():
    global df
    if df is not None:
        before = df.shape[0]
        df.drop_duplicates(inplace=True)
        after = df.shape[0]
        print(f"Removed {before - after} duplicate rows.")
    else:
        print("Load data first.")
```

# Handle Missing values

```python
def handle_missing():
    global df
    
    if df is None:
        print("Load data first.")
        return
    
    print("\n1. Drop rows")
    print("2. Fill with mean")
    print("3. Fill with median")
    print("4. Fill with custom value")
    
    choice = input("Choose option: ")
    
    if choice == "1":
        df.dropna(inplace=True)
        print("Rows with missing values removed.")
        
    elif choice == "2":
        df.fillna(df.mean(numeric_only=True), inplace=True)
        print("Missing values filled with mean.")
        
    elif choice == "3":
        df.fillna(df.median(numeric_only=True), inplace=True)
        print("Missing values filled with median.")
        
    elif choice == "4":
        value = input("Enter value: ")
        df.fillna(value, inplace=True)
        print("Missing values filled.")
        
    else:
        print("Invalid choice.")
```

# Basic Analysis

```python
def analyze_data():
    global df
    
    if df is None:
        print("Load data first.")
        return
    
    print("\n--- Basic Info ---")
    print("Total records:", len(df))
    print("\nColumn Summary:")
    print(df.info())
    
    print("\n--- Statistics ---")
    print(df.describe())
```

# Group Data

```python
def group_data():
    global df
    
    if df is None:
        print("Load data first.")
        return
    
    col = input("Enter column to group by: ").lower()
    
    if col in df.columns:
        grouped = df.groupby(col).mean(numeric_only=True)
        print(grouped)
    else:
        print("Column not found.")
```

# Filter Data

```python

def filter_data():
    global df
    
    if df is None:
        print("Load data first.")
        return
    
    col = input("Enter column name: ").lower()
    
    if col not in df.columns:
        print("Column not found.")
        return
    
    value = input("Enter value to filter: ").lower()
    
    filtered = df[df[col].astype(str).str.lower().str.contains(value)]
    
    print(filtered)
```

# Save Output

```python
def save_output():
    global df
    
    if df is None:
        print("No data to save.")
        return
    
    file_name = input("Enter output file name (without extension): ")
    
    try:
        df.to_excel(file_name + ".xlsx", index=False)
        df.to_csv(file_name + ".csv", index=False)
        print("Data saved successfully!")
    except Exception as e:
        print("Error:", e)
```

# Generate Report

```python
def generate_report():
    global df
    
    if df is None:
        print("Load data first.")
        return
    
    with open("report.txt", "w") as f:
        f.write("DATA SUMMARY REPORT\n\n")
        f.write(f"Total Records: {len(df)}\n\n")
        f.write("Columns:\n")
        f.write(str(df.columns.tolist()) + "\n\n")
        f.write("Statistics:\n")
        f.write(str(df.describe()))
    
    print("Report generated as report.txt")
```

# Menu System

```python
def menu():
    while True:
        print("\n===== DATA TOOL MENU =====")
        print("1. Load File")
        print("2. Clean Data")
        print("3. Analyze Data")
        print("4. Group Data")
        print("5. Filter/Search")
        print("6. Save Output")
        print("7. Generate Report")
        print("8. Exit")
        
        choice = input("Enter choice: ")
        
        if choice == "1":
            load_file()
            
        elif choice == "2":
            standardize_columns()
            remove_duplicates()
            handle_missing()
            
        elif choice == "3":
            analyze_data()
            
        elif choice == "4":
            group_data()
            
        elif choice == "5":
            filter_data()
            
        elif choice == "6":
            save_output()
            
        elif choice == "7":
            generate_report()
            
        elif choice == "8":
            print("Exiting...")
            break
        
        else:
            print("Invalid choice.")
```

# Run The Program

```python
menu()
```
