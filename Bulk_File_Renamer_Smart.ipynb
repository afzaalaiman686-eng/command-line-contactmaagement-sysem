# Import Required Libraries

```python
import os
import shutil
from datetime import datetime
```

# Select Folder 

```python
def select_folder():
    path = input("Enter folder path: ")
    if os.path.exists(path):
        return path
    else:
        print("Invalid path!")
        return None
```

# Get Files List

```python
def get_files(folder):
    files = []
    for file in os.listdir(folder):
        full_path = os.path.join(folder, file)
        if os.path.isfile(full_path):
            files.append(file)
    return files
```

# Rename Logic

```python
def rename_files(folder, prefix="", suffix="", replace_word=None, new_word=None):
    files = get_files(folder)
    renamed = []

    for i, file in enumerate(files, start=1):
        name, ext = os.path.splitext(file)

        if replace_word:
            name = name.replace(replace_word, new_word)

        new_name = f"{prefix}{name}_{i}{suffix}{ext}"
        renamed.append((file, new_name))

    return renamed
```

# Preview Changes

```python
def preview_changes(changes):
    print("\nPreview:")
    for old, new in changes:
        print(f"{old}  -->  {new}")
```

# Apply Changes Safely

```python
def apply_changes(folder, changes):
    for old, new in changes:
        old_path = os.path.join(folder, old)
        new_path = os.path.join(folder, new)

        if os.path.exists(new_path):
            print(f"Skipping (already exists): {new}")
            continue

        os.rename(old_path, new_path)

    print("Renaming completed.")
```

# Organize Files By Extensions

```python
def organize_files(folder):
    files = get_files(folder)

    for file in files:
        ext = file.split('.')[-1].upper()
        target_folder = os.path.join(folder, ext)

        if not os.path.exists(target_folder):
            os.makedirs(target_folder)

        src = os.path.join(folder, file)
        dest = os.path.join(target_folder, file)

        shutil.move(src, dest)

    print("Files organized by type.")
```

# Logging System

```python
def log_changes(changes):
    with open("log.txt", "a") as f:
        for old, new in changes:
            time = datetime.now()
            f.write(f"{old},{new},{time}\n")
```

# Menu System

```python
def menu():
    folder = None
    changes = []

    while True:
        print("\n1. Select Folder")
        print("2. Rename Files")
        print("3. Preview Changes")
        print("4. Apply Changes")
        print("5. Organize Files")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            folder = select_folder()

        elif choice == "2":
            if folder:
                prefix = input("Prefix: ")
                suffix = input("Suffix: ")
                changes = rename_files(folder, prefix, suffix)

        elif choice == "3":
            preview_changes(changes)

        elif choice == "4":
            apply_changes(folder, changes)
            log_changes(changes)

        elif choice == "5":
            organize_files(folder)

        elif choice == "6":
            break
```
