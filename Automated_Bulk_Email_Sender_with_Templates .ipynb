# Import Libraries

```python
# For sending emails
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

# For data handling
import csv
import json
import time
from datetime import datetime

# For file handling
import os
```

# Global configuraion(SMTP Setup)

```python
# Gmail SMTP configuration (you can change if using other provider)
SMTP_SERVER = "smtp.gmail.com"
SMTP_PORT = 587

# File to store email history
HISTORY_FILE = "email_history.json"
```

# Load Contacts Fom CSV/JSON

```python
def load_contacts(file_path):
    """
    Load contacts from CSV or JSON file.
    Expected fields: name, email, company
    """
    contacts = []

    try:
        if file_path.endswith(".csv"):
            with open(file_path, mode='r') as file:
                reader = csv.DictReader(file)
                for row in reader:
                    contacts.append(row)

        elif file_path.endswith(".json"):
            with open(file_path, mode='r') as file:
                contacts = json.load(file)

        else:
            print(" Unsupported file format. Use CSV or JSON only.")

    except Exception as e:
        print(" Error loading contacts:", e)

    return contacts
```

# Email Templates System

```python
# Predefined templates
templates = {
    "welcome": "Hello {name},\n\nWelcome to {company}! We are happy to have you onboard.\n\nRegards,\nTeam",
    
    "promotion": "Hi {name},\n\nWe have an exciting offer for {company} just for you!\nStay tuned.\n\nBest,\nMarketing Team"
}

def get_template(template_name):
    """Return selected template"""
    return templates.get(template_name, "")
```

# Send Email Function

```python
def send_email(sender_email, password, recipient_email, subject, body):
    """
    Sends an email using SMTP
    """

    try:
        # Create email object
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = recipient_email
        msg['Subject'] = subject

        # Attach body
        msg.attach(MIMEText(body, 'plain'))

        # Connect to server
        server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
        server.starttls()

        # Login
        server.login(sender_email, password)

        # Send email
        server.send_message(msg)
        server.quit()

        print(f" Email sent to {recipient_email}")

        return "Sent"

    except Exception as e:
        print(f" Failed for {recipient_email}:", e)
        return "Failed"
```

# Save Email History

```python
def save_history(record):
    """
    Save email sending history in JSON file
    """

    history = []

    # Load existing history if file exists
    if os.path.exists(HISTORY_FILE):
        with open(HISTORY_FILE, "r") as file:
            history = json.load(file)

    # Add new record
    history.append(record)

    # Save back to file
    with open(HISTORY_FILE, "w") as file:
        json.dump(history, file, indent=4)
```

```python
import csv

data = [
    ["name", "email", "company"],
    ["Ali", "ali@gmail.com", "HK Academy"],
    ["Sara", "sara@gmail.com", "Tech Corp"],
    ["Ahmed", "ahmed@gmail.com", "Soft Solutions"]
]

with open("contacts.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("CSV file created successfully!")
```

# Bulk Email Sender Function

```python
def send_bulk_emails(sender_email, password, contacts, template_name, subject):
    """
    Send emails to all contacts using selected template
    """

    template = get_template(template_name)

    if not template:
        print(" Template not found!")
        return

    for user in contacts:

        # Personalize email
        message = template.format(
            name=user.get("name", ""),
            company=user.get("company", "")
        )

        status = send_email(
            sender_email,
            password,
            user.get("email"),
            subject,
            message
        )

        # Save history
        record = {
            "email": user.get("email"),
            "subject": subject,
            "status": status,
            "timestamp": str(datetime.now())
        }

        save_history(record)

        # small delay (to avoid spam blocking)
        time.sleep(1)
```

# View Email History

```python
def view_history():
    """
    Display sent email history
    """

    if not os.path.exists(HISTORY_FILE):
        print("No history found.")
        return

    with open(HISTORY_FILE, "r") as file:
        history = json.load(file)

    for record in history:
        print(record)
```

# Simple Menu System (CLI)

```python
def menu():
    """
    Simple CLI menu for system control
    """

    contacts = []

    while True:
        print("\n====== EMAIL SENDER SYSTEM ======")
        print("1. Load Contacts")
        print("2. Send Bulk Emails")
        print("3. View History")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            path = input("Enter file path (CSV/JSON): ")
            contacts = load_contacts(path)
            print(f"Loaded {len(contacts)} contacts")

        elif choice == "2":
            sender = input("Enter your email: ")
            password = input("Enter app password: ")
            template = input("Enter template name (welcome/promotion): ")
            subject = input("Enter subject: ")

            send_bulk_emails(sender, password, contacts, template, subject)

        elif choice == "3":
            view_history()

        elif choice == "4":
            print("Exiting...")
            break

        else:
            print("Invalid option")
```

# Run  The Program

```python
## Start the system
menu()
```
