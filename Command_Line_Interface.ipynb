```python
import json
import os
import re

FILE_NAME = "contacts.json"


# ------------------ INITIAL DATA ------------------

DEFAULT_CONTACTS = [
    {
        "id": 1,
        "name": "Ali Ahmed",
        "phone": "03001234567",
        "email": "ali@gmail.com",
        "city": "Islamabad",
        "company": "TechSoft"
    },
    {
        "id": 2,
        "name": "Sara Khan",
        "phone": "03111234567",
        "email": "sara@yahoo.com",
        "city": "Lahore",
        "company": "DesignHub"
    },
    {
        "id": 3,
        "name": "Usman Tariq",
        "phone": "03211234567",
        "email": "usman@hotmail.com",
        "city": "Karachi",
        "company": "DevWorks"
    },
    {
        "id": 4,
        "name": "Ayesha Malik",
        "phone": "03331234567",
        "email": "ayesha@gmail.com",
        "city": "Peshawar",
        "company": "SoftNet"
    }
]


# ------------------ FILE HANDLING ------------------

def load_contacts():
    if not os.path.exists(FILE_NAME):
        return DEFAULT_CONTACTS

    try:
        with open(FILE_NAME, "r") as file:
            data = json.load(file)
            return data if data else DEFAULT_CONTACTS
    except:
        return DEFAULT_CONTACTS


def save_contacts(data):
    with open(FILE_NAME, "w") as file:
        json.dump(data, file, indent=4)


# ------------------ VALIDATION ------------------

def is_valid_email(email):
    return re.match(r"[^@]+@[^@]+\.[^@]+", email)


def is_valid_phone(phone):
    return phone.isdigit() and len(phone) >= 10


# ------------------ FEATURES ------------------

def view_contacts(contact_list):
    print("\n--- CONTACT LIST ---\n")

    if not contact_list:
        print("No contacts found.")
        return

    print(f"{'ID':5} {'Name':15} {'Phone':12} {'Email':25} {'City':12} {'Company'}")
    print("-" * 90)

    for c in contact_list:
        print(f"{c['id']:5} {c['name'][:15]:15} {c['phone']:12} {c['email'][:25]:25} {c['city'][:12]:12} {c['company']}")


def add_contact(contact_list):
    print("\n--- ADD CONTACT ---")

    name = input("Name: ").strip()
    phone = input("Phone: ").strip()
    email = input("Email: ").strip()
    city = input("City: ").strip()
    company = input("Company: ").strip()

    if not all([name, phone, email, city, company]):
        print(" All fields required!")
        return

    if not is_valid_phone(phone):
        print(" Invalid phone!")
        return

    if not is_valid_email(email):
        print(" Invalid email!")
        return

    new_id = max([c["id"] for c in contact_list], default=0) + 1

    contact_list.append({
        "id": new_id,
        "name": name,
        "phone": phone,
        "email": email,
        "city": city,
        "company": company
    })

    save_contacts(contact_list)
    print("✅ Contact added!")


def search_contacts(contact_list):
    print("\n--- SEARCH ---")
    query = input("Enter name/phone/email: ").lower()

    results = [
        c for c in contact_list
        if query in c["name"].lower()
        or query in c["phone"]
        or query in c["email"].lower()
    ]

    view_contacts(results)


def filter_contacts(contact_list):
    print("\n--- FILTER ---")
    print("1. City")
    print("2. Company")

    choice = input("Choose: ")

    if choice == "1":
        city = input("City: ").lower()
        results = [c for c in contact_list if c["city"].lower() == city]

    elif choice == "2":
        company = input("Company: ").lower()
        results = [c for c in contact_list if c["company"].lower() == company]

    else:
        print(" Invalid choice")
        return

    view_contacts(results)


def update_contact(contact_list):
    print("\n--- UPDATE ---")

    try:
        cid = int(input("Enter ID: "))
    except:
        print(" Invalid ID")
        return

    for c in contact_list:
        if c["id"] == cid:

            name = input("New name: ")
            phone = input("New phone: ")
            email = input("New email: ")
            city = input("New city: ")
            company = input("New company: ")

            if name:
                c["name"] = name

            if phone and is_valid_phone(phone):
                c["phone"] = phone

            if email and is_valid_email(email):
                c["email"] = email

            if city:
                c["city"] = city

            if company:
                c["company"] = company

            save_contacts(contact_list)
            print("✅ Updated!")
            return

    print(" Not found")


def delete_contact(contact_list):
    print("\n--- DELETE ---")

    try:
        cid = int(input("Enter ID: "))
    except:
        print(" Invalid ID")
        return

    new_list = [c for c in contact_list if c["id"] != cid]

    if len(new_list) == len(contact_list):
        print(" Not found")
    else:
        contact_list.clear()
        contact_list.extend(new_list)
        save_contacts(contact_list)
        print("🗑 Deleted!")


# ------------------ MENU ------------------

def menu():
    contacts = load_contacts()

    while True:
        print("\n===== CONTACT SYSTEM =====")
        print("1. View Contacts")
        print("2. Add Contact")
        print("3. Search")
        print("4. Filter")
        print("5. Update")
        print("6. Delete")
        print("7. Exit")

        choice = input("Choose: ")

        if choice == "1":
            view_contacts(contacts)
        elif choice == "2":
            add_contact(contacts)
        elif choice == "3":
            search_contacts(contacts)
        elif choice == "4":
            filter_contacts(contacts)
        elif choice == "5":
            update_contact(contacts)
        elif choice == "6":
            delete_contact(contacts)
        elif choice == "7":
            print("Goodbye!")
            break
        else:
            print(" Invalid option")


# ------------------ RUN ------------------

if __name__ == "__main__":
    menu()
```
