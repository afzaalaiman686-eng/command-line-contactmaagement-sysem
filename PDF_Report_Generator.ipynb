# Import Required Libraries

```python
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet
from datetime import datetime
import os
```

# Create Folder for Reports 

```python
# Create folder if it doesn't exist
import os

os.makedirs("reports", exist_ok=True)
```

# Function: Add Manual Data

```python

def add_manual_data():
    data = []

    print("Enter student/company records (type 'done' to stop)\n")

    while True:
        name = input("Name: ")
        if name.lower() == "done":
            break

        record = {
            "name": name,
            "id": input("ID: "),
            "email": input("Email: "),
            "role": input("Course / Department / Role: "),
            "details": input("Performance / Details: ")
        }

        data.append(record)
        print(" Record added\n")

    return data
```

# Function Generate PDF Report

```python
def generate_pdf(data, report_type="Student"):
    if not data:
        print("No data available to generate report.")
        return

    filename = f"reports/{report_type}_Report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.pdf"
    pdf = SimpleDocTemplate(filename)

    styles = getSampleStyleSheet()
    elements = []

    # Title
    title = Paragraph(f"<b>{report_type} Report</b>", styles["Title"])
    elements.append(title)
    elements.append(Spacer(1, 12))

    # Date
    date_info = Paragraph(f"Generated on: {datetime.now()}", styles["Normal"])
    elements.append(date_info)
    elements.append(Spacer(1, 12))

    # Table Header
    table_data = [["Name", "ID", "Email", "Role", "Details"]]

    # Add rows
    for item in data:
        table_data.append([
            item["name"],
            item["id"],
            item["email"],
            item["role"],
            item["details"]
        ])

    # Table Styling
    table = Table(table_data, colWidths=100)

    table.setStyle(TableStyle([
        ("BACKGROUND", (0, 0), (-1, 0), colors.grey),
        ("TEXTCOLOR", (0, 0), (-1, 0), colors.whitesmoke),
        ("ALIGN", (0, 0), (-1, -1), "CENTER"),
        ("GRID", (0, 0), (-1, -1), 0.5, colors.black),
        ("BACKGROUND", (0, 1), (-1, -1), colors.beige),
    ]))

    elements.append(table)

    pdf.build(elements)

    print(f"\n PDF Successfully Generated: {filename}")
```

# Menu System

```python
data_store = []

while True:
    print("\n===== PDF REPORT GENERATOR =====")
    print("1. Add Data Manually")
    print("2. Generate Student Report")
    print("3. Generate Company Report")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        data_store = add_manual_data()

    elif choice == "2":
        generate_pdf(data_store, "Student")

    elif choice == "3":
        generate_pdf(data_store, "Company")

    elif choice == "4":
        print("Exiting program...")
        break

    else:
        print("nvalid choice. Try again.")
```
