# Introduction
In this project, we are building a CLI-based Python tool that can encrypt and decrypt user data using a custom algorithm. The program also supports saving encrypted data to a file and loading it back for decryption.
The encryption method used here is:
- Character shifting.
- Reversing the string.
- Key-based transformation (password adds extra security).

# Import Required Libraries 

```python
import json
import datetime
```

# Custom Encryption Function

```python
def encrypt_text(text, key):
    """
    This function encrypts the given text using:
    1. Character shifting
    2. String reversal
    3. Key-based transformation
    """
    try:
        encrypted = ""
        
        # Step 1: Shift characters using key
        for i, char in enumerate(text):
            shift = ord(key[i % len(key)])  # take key character ASCII
            encrypted_char = chr((ord(char) + shift) % 256)
            encrypted += encrypted_char
        
        # Step 2: Reverse the string
        encrypted = encrypted[::-1]
        
        return encrypted
    
    except Exception as e:
        print("Error during encryption:", e)
```

# Custom Decryption Function

```python
def decrypt_text(encrypted_text, key):
    """
    This function reverses the encryption process
    """
    try:
        # Step 1: Reverse back
        encrypted_text = encrypted_text[::-1]
        
        decrypted = ""
        
        # Step 2: Reverse character shifting
        for i, char in enumerate(encrypted_text):
            shift = ord(key[i % len(key)])
            decrypted_char = chr((ord(char) - shift) % 256)
            decrypted += decrypted_char
        
        return decrypted
    
    except Exception as e:
        print("Error during decryption:", e)
```

# Save Encrypted data into File

```python
def save_to_file(data):
    """
    Saves encrypted data into a JSON file with timestamp
    """
    try:
        file_data = {
            "timestamp": str(datetime.datetime.now()),
            "encrypted_data": data
        }
        
        with open("encrypted_data.json", "w") as file:
            json.dump(file_data, file)
        
        print("Data saved successfully!")
    
    except Exception as e:
        print("Error saving file:", e)
```

# Load Encrypted Data From File

```python
def load_from_file():
    """
    Loads encrypted data from file
    """
    try:
        with open("encrypted_data.json", "r") as file:
            data = json.load(file)
        
        print("Data loaded successfully!")
        return data["encrypted_data"]
    
    except FileNotFoundError:
        print("File not found!")
    except Exception as e:
        print("Error loading file:", e)
```

# Main Menu System (CLI)

```python
def main():
    encrypted_data = ""
    
    while True:
        print("\n====== Encryption Tool Menu ======")
        print("1. Encrypt Data")
        print("2. Decrypt Data")
        print("3. Save Data")
        print("4. Load Data")
        print("5. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            text = input("Enter text to encrypt: ")
            key = input("Enter encryption key: ")
            
            encrypted_data = encrypt_text(text, key)
            print("Encrypted Data:", encrypted_data)
        
        elif choice == "2":
            if not encrypted_data:
                print("No encrypted data available!")
            else:
                key = input("Enter decryption key: ")
                decrypted = decrypt_text(encrypted_data, key)
                print("Decrypted Data:", decrypted)
        
        elif choice == "3":
            if encrypted_data:
                save_to_file(encrypted_data)
            else:
                print("No data to save!")
        
        elif choice == "4":
            encrypted_data = load_from_file()
            print("Encrypted Data:", encrypted_data)
        
        elif choice == "5":
            print("Exiting program...")
            break
        
        else:
            print("Invalid choice! Try again.")
```

# Run The Program

```python
### if __name__ == "__main__":
main()
```
