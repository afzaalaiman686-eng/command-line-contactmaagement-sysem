# Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import csv
import json
```

# Add Headers 

```python
def get_headers():
    return {
        "User-Agent": "Mozilla/5.0"
    }
```

# Fetch Webpage

```python
def fetch_page(url):
    try:
        response = requests.get(url, headers=get_headers())
        response.raise_for_status()
        return response.text
    except requests.exceptions.RequestException as e:
        print("Error fetching page:", e)
        return None
```

# Parse Data  

```python
def parse_data(html):
    soup = BeautifulSoup(html, "html.parser")
    data = []

    quotes = soup.find_all("div", class_="quote")

    for q in quotes:
        title = q.find("span", class_="text").text
        author = q.find("small", class_="author").text

        item = {
            "title": title,
            "author": author
        }

        data.append(item)

    return data
```

# Pagination

```python
def scrape_multiple_pages(base_url, pages=2):
    all_data = []

    for i in range(1, pages + 1):
        url = f"{base_url}/page/{i}/"
        print("Scraping:", url)

        html = fetch_page(url)
        if html:
            page_data = parse_data(html)
            all_data.extend(page_data)

    return all_data
```

# Search Function

```python
def search_data(data, keyword):
    results = []

    for item in data:
        if keyword.lower() in item["title"].lower():
            results.append(item)

    return results
```

# Save to JSON

```python
def save_json(data, filename="data.json"):
    with open(filename, "w") as f:
        json.dump(data, f, indent=4)
    print("Saved to JSON")
    
```

# Save to CSV

```python
def save_csv(data, filename="data.csv"):
    if not data:
        return

    keys = data[0].keys()

    with open(filename, "w", newline='', encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=keys)
        writer.writeheader()
        writer.writerows(data)

    print("Saved to CSV")
```

# Menu System

```python
def menu():
    data = []

    while True:
        print("\n1. Scrape Data")
        print("2. Search")
        print("3. Save JSON")
        print("4. Save CSV")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            url = "https://quotes.toscrape.com"
            data = scrape_multiple_pages(url, pages=3)
            print("Scraped", len(data), "items")

        elif choice == "2":
            keyword = input("Enter keyword: ")
            results = search_data(data, keyword)
            print("Found:", len(results))
            for r in results[:5]:
                print(r)

        elif choice == "3":
            save_json(data)

        elif choice == "4":
            save_csv(data)

        elif choice == "5":
            break
```

# Run Program

```python
menu()
```
