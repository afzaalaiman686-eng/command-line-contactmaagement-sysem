# Introduction
This project is a CLI-based Python tool that reads log files and analyzes them to find errors, warnings, and info messages. It also allows searching logs and generating reports.

# Import Libraries

```python
import re   # used for pattern matching (regex)
import csv
```

# Load Log File

```python
def load_log_file():
    """
    This function loads a log file and reads it line by line
    """
    try:
        file_path = input("Enter log file path (.txt or .log): ")
        
        with open(file_path, "r") as file:
            logs = file.readlines()
        
        print("File loaded successfully!")
        return logs
    
    except FileNotFoundError:
        print("File not found! Please check the path.")
        return []
    
    except Exception as e:
        print("Error loading file:", e)
        return []
```

# Analyze Logs

```python
def analyze_logs(logs):
    """
    This function analyzes logs and counts ERROR, WARNING, INFO
    """
    error_count = 0
    warning_count = 0
    info_count = 0
    
    timestamps = []

    for line in logs:
        # Case-insensitive matching
        if re.search(r'ERROR', line, re.IGNORECASE):
            error_count += 1
        elif re.search(r'WARNING', line, re.IGNORECASE):
            warning_count += 1
        elif re.search(r'INFO', line, re.IGNORECASE):
            info_count += 1
        
        # Extract timestamp (simple format example)
        match = re.search(r'\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}', line)
        if match:
            timestamps.append(match.group())
    
    result = {
        "total": len(logs),
        "errors": error_count,
        "warnings": warning_count,
        "info": info_count,
        "timestamps": timestamps
    }
    
    print("Logs analyzed successfully!")
    return result
```

# Search Logs

```python
def search_logs(logs):
    """
    Search logs by keyword (case-insensitive)
    """
    keyword = input("Enter keyword to search: ")
    
    results = []
    
    for line in logs:
        if keyword.lower() in line.lower():
            results.append(line)
    
    if results:
        print("\nSearch Results:")
        for r in results:
            print(r.strip())
    else:
        print("No matching logs found.")
```

# Filter Logs

```python
def filter_logs(logs):
    """
    Filter logs by level (ERROR/WARNING/INFO)
    """
    level = input("Enter level (ERROR/WARNING/INFO): ").upper()
    
    filtered = []
    
    for line in logs:
        if level in line.upper():
            filtered.append(line)
    
    if filtered:
        print("\nFiltered Logs:")
        for f in filtered:
            print(f.strip())
    else:
        print("No logs found for this level.")
```

# Generate Reports 

```python
def generate_report(result):
    """
    Generate summary report and save to file
    """
    print("\n------ Log Summary Report ------")
    print("Total Logs:", result["total"])
    print("Errors:", result["errors"])
    print("Warnings:", result["warnings"])
    print("Info:", result["info"])
    
    choice = input("Save report? (y/n): ")
    
    if choice.lower() == "y":
        try:
            # Save as TXT
            with open("log_report.txt", "w") as file:
                file.write("Log Report\n")
                file.write(f"Total Logs: {result['total']}\n")
                file.write(f"Errors: {result['errors']}\n")
                file.write(f"Warnings: {result['warnings']}\n")
                file.write(f"Info: {result['info']}\n")
            
            # Save as CSV
            with open("log_report.csv", "w", newline="") as file:
                writer = csv.writer(file)
                writer.writerow(["Type", "Count"])
                writer.writerow(["Total", result["total"]])
                writer.writerow(["Errors", result["errors"]])
                writer.writerow(["Warnings", result["warnings"]])
                writer.writerow(["Info", result["info"]])
            
            print("Report saved successfully!")
        
        except Exception as e:
            print("Error saving report:", e)
```

# Main Menu System

```python
def main():
    global logs
    logs = []
    result = None
    
    while True:
        print("\n====== Log Analyzer Menu ======")
        print("1. Load Log File")
        print("2. Analyze Logs")
        print("3. Search Logs")
        print("4. Filter Logs")
        print("5. Generate Report")
        print("6. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            logs = load_log_file()
            print ("logs stored",len(logs))
        
        elif choice == "2":
            if logs:
                result = analyze_logs(logs)
            else:
                print("Load file first!")
        
        elif choice == "3":
            if logs:
                search_logs(logs)
            else:
                print("Load file first!")
        
        elif choice == "4":
            if logs:
                filter_logs(logs)
            else:
                print("Load file first!")
        
        elif choice == "5":
            if result:
                generate_report(result)
            else:
                print("Analyze logs first!")
        
        elif choice == "6":
            print("Exiting program...")
            break
        
        else:
            print("Invalid choice!")
```

# Run The program

```python
if __name__ == "__main__":
    main()
```
