# Purpose of the Project
- The main purpose of this project is to understand and implement the core concepts of user authentication systems using Python
- It focuses on:
- Learning how real-world login/signup systems work
- Implementing secure password storage using hashing
- Practicing file handling with JSON
- Applying input validation and error handling
- Building modular and structured programs
- This project also helps in developing problem-solving skills and understanding basic cybersecurity practices

```python
import json
import os
import hashlib
import re

FILE_NAME = "users.json"

# -----------------------------
# Utility Functions
# -----------------------------

def load_users():
    """Load users from JSON file"""
    if not os.path.exists(FILE_NAME):
        return {}
    
    with open(FILE_NAME, "r") as file:
        try:
            return json.load(file)
        except json.JSONDecodeError:
            return {}


def save_users(users):
    """Save users to JSON file"""
    with open(FILE_NAME, "w") as file:
        json.dump(users, file, indent=4)


def hash_password(password):
    """Hash password using SHA-256"""
    return hashlib.sha256(password.encode()).hexdigest()


def is_valid_email(email):
    """Check email format using regex"""
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    return re.match(pattern, email)


def is_strong_password(password):
    """
    Password rules:
    - At least 8 characters
    - Contains uppercase, lowercase, digit
    """
    if len(password) < 8:
        return False
    if not re.search(r"[A-Z]", password):
        return False
    if not re.search(r"[a-z]", password):
        return False
    if not re.search(r"\d", password):
        return False
    return True


# -----------------------------
# Core Features
# -----------------------------

def register(users):
    print("\n--- Register ---")
    
    username = input("Enter username: ").strip()
    email = input("Enter email: ").strip()
    password = input("Enter password: ").strip()

    # Check duplicates
    for user in users.values():
        if user["email"] == email:
            print("Email already exists!")
            return
    
    if username in users:
        print("Username already taken!")
        return

    # Validate email
    if not is_valid_email(email):
        print("Invalid email format!")
        return

    # Validate password
    if not is_strong_password(password):
        print("Weak password! Must be 8+ chars, include upper, lower, number.")
        return

    # Store user
    users[username] = {
        "email": email,
        "password": hash_password(password),
        "role": "user"
    }

    save_users(users)
    print("Registration successful!")


def login(users):
    print("\n--- Login ---")
    
    identifier = input("Enter username or email: ").strip()
    password = input("Enter password: ").strip()

    hashed = hash_password(password)

    # Search user
    for username, data in users.items():
        if identifier == username or identifier == data["email"]:
            if data["password"] == hashed:
                print(f"Login successful! Welcome {username}")
                return username
            else:
                print("Incorrect password!")
                return None

    print("User not found!")
    return None


def reset_password(users):
    print("\n--- Reset Password ---")
    
    username = input("Enter username: ").strip()
    
    if username not in users:
        print("User not found!")
        return

    new_password = input("Enter new password: ").strip()

    if not is_strong_password(new_password):
        print("Weak password!")
        return

    users[username]["password"] = hash_password(new_password)
    save_users(users)
    print("Password updated successfully!")


def delete_account(users):
    print("\n--- Delete Account ---")
    
    username = input("Enter username: ").strip()
    
    if username not in users:
        print("User not found!")
        return

    confirm = input("Are you sure? (yes/no): ").lower()

    if confirm == "yes":
        del users[username]
        save_users(users)
        print("Account deleted.")
    else:
        print("Cancelled.")


# -----------------------------
# Main Menu
# -----------------------------

def main():
    users = load_users()

    while True:
        print("\n====== MENU ======")
        print("1. Register")
        print("2. Login")
        print("3. Reset Password")
        print("4. Delete Account")
        print("5. Exit")

        choice = input("Choose option: ")

        if choice == "1":
            register(users)
        elif choice == "2":
            login(users)
        elif choice == "3":
            reset_password(users)
        elif choice == "4":
            delete_account(users)
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid choice!")


# Run program
if __name__ == "__main__":
    main()
```
